```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('data.csv')
```

## FINDING AND REMOVING OUTLIERS

```python
## kills columns
```

```python
check_otlr = df.apply(lambda x: True if x['kills'] > 30  else False , axis=1)
no_of_otliers = len(check_otlr[check_otlr == True].index)
no_of_otliers
```

```python
df.drop(df[ df['kills'] > 30 ].index , inplace=True)
df.shape
```

```python
## matchDuration
```

```python
check_otlr = df.apply(lambda x: True if x['matchDuration'] < 500  else False , axis=1)
no_of_otliers = len(check_otlr[check_otlr == True].index)
no_of_otliers
```

```python
df.drop(df[ df['matchDuration'] < 500 ].index , inplace=True)
df.shape
```

```python
## rankPoints
```

```python
check_otlr = df.apply(lambda x: True if x['rankPoints'] > 3300  else False , axis=1)
no_of_otliers = len(check_otlr[check_otlr == True].index)
no_of_otliers
```

```python
df.drop(df[ df['rankPoints'] > 3300 ].index , inplace=True)
df.shape
```

```python
df.to_csv('data kills outliers',index=False,header=True)
```
