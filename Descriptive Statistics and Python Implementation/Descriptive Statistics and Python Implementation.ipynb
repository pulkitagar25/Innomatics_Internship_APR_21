## Mean:--

The mean is equal to the sum of all the values in the data set divided by the number of values in the data set i.e the calculated average. It susceptible to outliers when unusual values are added it gets skewed i.e deviates from the typical central value.






```python
## WITHOUT LIBRARY FUNCTION

print("Without using built-in function \n")

def mean_filling(df):
    
    a=sum(df['Mthly_HH_Income'])/len(df['Mthly_HH_Income'])
    return a;
    
print("Mean is:- ",mean_filling(df))
```

## MEDIAN:--

The median is the middle value for a dataset that has been arranged in order of magnitude. Median is a better alternative to mean as it is less affected by outliers and skewness of the data. The median value is much closer than the typical central value.

If the total number of values is odd then

MEDIAN = (n+1)/2 -th term

If the total number of values is odd then

MEDIAN = average of the n/2 -th and the (n/2 + 1) -th terms





```python
## WITHOUT LIBRARY FUNCTION:--

print("Without using built-in function \n")

def median(df):

    q, r = divmod(len(df), 2)
    
    if r:
    
        return sorted(df)[q]
        
    return sum(sorted(df)[q - 1:q + 1]) / 2
    

print("Median is:- :", median(df['Mthly_HH_Income']))
```

```python
## WITH LIBRARY FUNCTION:--

print("Using built-in function \n")



med = df['Mthly_HH_Income'].median()


print("Median is:- ", med)


print("\n")
```

## LaTEX FORM

### FOR EVEN MEDIAN= \text { Median }=\frac{\left(\frac{n}{2}\right)^{t h}+\left(\frac{n}{2}+1\right)^{t h}}{2}

FOR ODD:--
    
$$ \text { Median }=\left(\frac{n+1}{2}\right)^{t h} $$

FOR EVEN:--

    
$$ \text { Median }=\frac{\left(\frac{n}{2}\right)^{t h}+\left(\frac{n}{2}+1\right)^{t h}}{2} $$

## Mode:--

The mode is the most commonly occurring value in the dataset. The mode can, therefore sometimes consider the mode as being the most popular option.

For Example, In a dataset containing {13,35,54,54,55,56,57,67,85,89,96} values. Mean is 60.09. Median is 56. Mode is 54





```python
## WITHOUT USING LIBRARY FUNCTION:--

print("Without using built-in function \n")

from collections import Counter

n = len(df['Mthly_HH_Expense'])  
data = Counter(df['Mthly_HH_Expense'])

get_mode = dict(data) 

mode = [k for k, v in get_mode.items() if v == max(list(data.values()))] 
  
if len(mode) == n:

    get_mode = "No mode found"
    
else: 

    get_mode = "Mode is : " + ', '.join(map(str, mode)) 
      
print(get_mode) 
```

## Variance:--

Variance measures how far is the sum of squared distances from each point to the mean i.e the dispersion around the mean.

Variance is the average of all squared deviations.



### Population variance:--

When you have collected data from every member of the population that youre interested in, you can get an exact value for population variance.

### Sample variance:--
When you collect data from a sample, the sample variance is used to make estimates or inferences about the population variance.


Variance matters for two main reasons:

1. Parametric statistical tests are sensitive to variance.

2. Comparing the variance of samples helps you assess group differences.


```python
## WITHOUT LIBRARY FUNCTION
print("Without using built-in function \n")

avg = sum(df['Mthly_HH_Income'])/len(df['Mthly_HH_Income'])
var = sum((x-avg)**2 for x in df['Mthly_HH_Income']) / (len(df['Mthly_HH_Income'])-1)
print("variance is:- ",round(var,0))
```

#### IF WE CALCULATE VARIANCE USING SIMPLE VARIANCE FORMULA WE GET:-- 681100853.0 

#### AND IF WE CALCULATE VARIANCE USING POPULATION VARIANCE FORMULA WE GET:-- 667478836.0. 

#### THIS DIFFERENCE IS CAUSED DUE TO (N-1) FACTOR IN SAMPLE VARIANCE FORMULA.

```python
## WITH LIBRARY FUNCTION

print("Using built-in function \n")

var = np.var(df['Mthly_HH_Income'])
print("variance is:- ",round(var,0))
```

### LaTEX FORM OF POPULATION VARIANCE:--  \sigma^{2}=\frac{\sum_{i=1}^{N}\left(x_{i}-\mu\right)^{2}}{N}
    
    
POPULATION VARIANCE: $$ \sigma^{2}=\frac{\sum_{i=1}^{N}\left(x_{i}-\mu\right)^{2}}{N} $$

### LaTEX FORM OF SAMPLE VARIANCE:--  s^{2}=\frac{\sum_{i=1}^{n}\left(x_{i}-\bar{x}\right)^{2}}{n-1}

SAMPLE VARIANCE: $$ s^{2}=\frac{\sum_{i=1}^{n}\left(x_{i}-\bar{x}\right)^{2}}{n-1} $$

## Standard Deviation:--

As Variance suffers from unit difference so standard deviation is used. The square root of the variance is the standard deviation. It tells about the concentration of the data around the mean of the data set.

Its symbol is  (the greek letter sigma)

The formula is easy: it is the square root of the Variance.




```python
## WITHOUT LIBRARY FUNCTION

print("Without using built-in function \n")


avg= round(sum(df['Mthly_HH_Income'])/len(df['Mthly_HH_Income']),0)

var = sum((x- avg)**2 for x in df['Mthly_HH_Income']) / (len(df['Mthly_HH_Income']) -1)
print("variance is:- ",round(var,0))

sd= var**0.5

print("Standard deviation is:- ",round(sd,0))
```

```python
## WITH LIBRARY FUNCTION

print("Using built-in function \n")

var = np.var(df['Mthly_HH_Income'])
print("variance is:- ",var)
std_dev = math.sqrt(var)

print("Standard deviation is:- ",std_dev)

```

#### LaTEX FORM:-- \mathrm{s}_{\mathrm{x}}=\sqrt{\frac{\sum_{\mathrm{i}=1}^{\mathrm{n}}\left(\mathrm{x}_{\mathrm{i}}-\overline{\mathrm{x}}\right)^{2}}{\mathrm{n}-1}}

STANDARD DEVIATION: $$ \mathrm{s}_{\mathrm{x}}=\sqrt{\frac{\sum_{\mathrm{i}=1}^{\mathrm{n}}\left(\mathrm{x}_{\mathrm{i}}-\overline{\mathrm{x}}\right)^{2}}{\mathrm{n}-1}} $$

## Correlation:--

Correlation gives a better understanding of covariance. It is normalized covariance. Correlation tells us how correlated the variables are to each other. It is also called as Pearson Correlation Coefficient.

The correlation coefficient is a value that indicates the strength of the relationship between variables. The coefficient can take any values from -1 to 1. The interpretations of the values are:

a) -1: Perfect negative correlation. The variables tend to move in opposite directions (i.e., when one variable increases, the other variable decreases).

b)  0: No correlation. The variables do not have a relationship with each other.

c)  1: Perfect positive correlation. The variables tend to move in the same direction (i.e., when one variable increases, the other variable also increases).



```python
print("Without using built-in function \n")

X = df['Mthly_HH_Income']
Y = df['Mthly_HH_Expense']

Xmean = sum(X)/len(X)
Ymean = sum(Y)/len(Y)

x = [var-Xmean for var in X]
y = [var-Ymean for var in Y]

xy =[a*b for a,b in list(zip(x,y))]
sum_xy = sum(xy)

x_square = [a*a for a in x]
y_square = [b*b for b in y]

sum_x_square = sum(x_square)
sum_y_square = sum(y_square)

sum_x_square_sum_y_square = sum_x_square*sum_y_square
sqrt_sum_x_square_sum_y_square = math.sqrt(sum_x_square_sum_y_square)

r = sum_xy/sqrt_sum_x_square_sum_y_square

print(r)
```

## NORMAL DISTRIBUTION:-- 

Frequency distributions are all shapes, but a statistically normal frequency distribution has the mean as its mean, its mode, and its median and it looks like a symmetric mound or bell.

The word normal is a very powerful adjective when used to describe a frequency distribution or when used to describe the data of a sample or population.

### Probability and A Normal Distribution:--

The number line is marked in terms of the mean and the standard deviation.

 
Since normally distributed data is spread in terms of the mean and standard deviation, the percent of scores, within the same region, remains the same from normal distribution to normal distribution.







### Properties of a normal distribution:-

a)  The mean, mode and median are all equal.

b)  The curve is symmetric at the center (i.e. around the mean, ).

c)  Exactly half of the values are to the left of center and exactly half the values are to the right.

d)  The total area under the curve is 1.

## Positively Skewed Normal Distribution:--

A right-skewed distribution has a long right tail. Right-skewed distributions are also called positive-skew distributions. Thats because there is a long tail in the positive direction on the number line. The mean is also to the right of the peak.

The general relationship among the central tendency measures in a positively skewed distribution may be expressed using the following inequality:

Mean  >  Median  >  Mode

## Negatively Skewed Normal Distribution:--

A left-skewed distribution has a long left tail. Left-skewed distributions are also called negatively-skewed distributions. Thats because there is a long tail in the negative direction on the number line. The mean is also to the left of the peak.

The general relationship between the central tendency measures in a negatively skewed distribution may be expressed using the following inequality:

Mode  >  Median  > Mean



Generally if the distribution of data is skewed to the left, the mean is less than the median, which is often less than the mode. If the distribution of data is skewed to the right, the mode is often less than the median, which is less than the mean.

## QQ PLOT:--

Q Q Plots (Quantile-Quantile plots) are plots of two quantiles against each other. A quantile is a fraction where certain values fall below that quantile. For example, the median is a quantile where 50% of the data fall below that point and 50% lie above it. The purpose of Q Q plots is to find out if two sets of data come from the same distribution. A 45 degree angle is plotted on the Q Q plot; if the two data sets come from a common distribution, the points will fall on that reference line.



```python
import statsmodels.api as sm 
import pylab as py 
  
data_points = np.random.normal(0, 1, 100)     
  
sm.qqplot(data_points, line ='45') 
py.show()
```
